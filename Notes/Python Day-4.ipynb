```python
# Calculator

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
print(f"The addition of {a} & {b} is: {a+b}")
print(f"The subtraction of {a} & {b} is: {a-b}")
print(f"The multiplication of {a} & {b} is: {a*b}")
print(f"The division of {a} & {b} is: {a/b}")
print(f"{a} raised to the power of {b} is: {a**b}")
```

```python
# mask ending digits of input number
# ask user to enter credit card number
#print the number with the last four digits masked with XXXX

output= "XXXX XXXX XXXX YYYY"

number = input("Enter credit card number: ")

# last_digits = number[-4]+number[-3]+number[-2]+number[-1] #XXXX XXXX XXXX 1234
last_digits = number[-4::] #XXXX XXXX XXXX 1234
final_word= output.replace("YYYY", last_digits)

print(final_word)
```
